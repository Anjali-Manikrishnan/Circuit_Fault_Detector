```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier

# Load the dataset (same as the training one)
df = pd.read_csv("circuit_fault_data.csv")

# Prepare the model
X = df[['Input Voltage (V)', 'Output Current (A)']]
y = df['Fault Type']

model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Take user input
try:
    vin = float(input("Enter Input Voltage (V): "))
    iout = float(input("Enter Output Current (A): "))

    # Predict
    prediction = model.predict([[vin, iout]])
    print(f"\nüîç The predicted fault type is: **{prediction[0]}**")

except ValueError:
    print("‚ùå Invalid input. Please enter numerical values.")
```
